```python
import pandas as pd

df = pd.read_csv("IMDB Dataset.csv")
df.head()
```

```python
def get_pos_tags(text):
    doc = nlp(text)
    return [(token.text, token.pos_) for token in doc]

# Apply on a small sample first (full dataset can be slow)
df_sample = df.head(5).copy()
df_sample['pos_tags'] = df_sample['review'].apply(get_pos_tags)
df_sample[['review', 'pos_tags']].head()
```

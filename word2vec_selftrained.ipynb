```python
import numpy as np
import pandas as pd
```

```python
import gensim
import os 
```

```python
from gensim.utils import simple_preprocess

story = []

# Read files line-by-line and tokenize each non-empty line
data_dir = 'got_books'
for filename in os.listdir(data_dir):
    path = os.path.join(data_dir, filename)
    # use encoding to avoid errors on some files
    with open(path, encoding='utf-8', errors='ignore') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            tokens = simple_preprocess(line)
            if tokens:
                story.append(tokens)
```

```python
story
```

```python
from gensim.models import Word2Vec
model = Word2Vec(
    window=10,
    min_count = 2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```

```python
model.wv.doesnt_match(['jon', 'rkion', 'arya', 'robb', 'sansa', 'bran'])
```

```python
import gensim.downloader as api
# Load a pretrained model via gensim downloader.
# 'word2vec-google-news-300' is large (~1.6GB). Use 'glove-wiki-gigaword-100' for a smaller test.
# model = api.load('word2vec-google-news-300')
model = api.load('glove-wiki-gigaword-100')
print('loaded model; vocab size:', len(model.key_to_index))
```
